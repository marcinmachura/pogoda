```python
import ast
climate =open("climate.tsv")
data = {}
ii=0
for line in climate:
    ii = ii +1
    if ii % 10000 == 0:
        print(ii)  
    cols = line.strip().split('\t')
    lat = round(float(cols[0]),3)
    lon = round(float(cols[1]),3)
    payload = ast.literal_eval(cols[2])        
    data[(lat,lon)] = payload
    
```

```python
import pickle
```

```python
with open("climate.pkl", "wb") as f:
    pickle.dump(data, f)
```

```python
import json
```

```python
with open("climate.json", "w") as f:
    json.dump(data, f)
```
