```python
import ast
import json
tg_data = open("outputTG.tsv")
TG = {}
ii = 0
for line in tg_data:
    ii = ii +1
    if ii % 10000 == 0:
        print(ii)
    tg = line.strip().split('\t')
    tg_lat = round(float(tg[0]),3)
    tg_lon = round(float(tg[1]),3)
    tgg = tg[2].replace('nan', '-101033')
    payload = ast.literal_eval(tgg)
    data = {}
    for year in payload.keys():
        if -101033 not in payload[year]:
            data[year] = [round(x,3) for x in payload[year]]
    TG[(tg_lat, tg_lon)] = data
    
```

```python
rr_data = open("outputRR.tsv")
RR = {}
ii = 0
for line in rr_data:
    ii = ii +1
    if ii % 10000 == 0:
        print(ii)
    rr = line.strip().split('\t')
    rr_lat = round(float(rr[0]),3)
    rr_lon = round(float(rr[1]),3)
    if (rr_lat, rr_lon) not in TG:
        continue
    rrr = rr[2].replace('nan', '-101033')
    payload = ast.literal_eval(rrr)
    data = {}
    for year in payload.keys():
        if -101033 not in payload[year]:
            data[year] = [round(x,3) for x in payload[year]]
    RR[(rr_lat, rr_lon)] = data
    
```

```python
len(RR)
```

```python
# cleaning up NaNs in TG
outF = open("climate.tsv","w")
ii = 0
for key in RR.keys():
    ii = ii+1
    if ii % 10000 == 0:
        print(ii)
    if key in TG:
        payload = {}
        for year in RR[key]:
            if year in TG[key]:
                data = zip(TG[key][year], RR[key][year])
                payload[year] = list(data)
        outF.write(f"{key[0]}\t{key[1]}\t{payload}\n")
        
```
