```python
from typing import *
import json
import numpy as np

def decodeCoordinates (encoded: np.int64) -> Tuple[np.float32,np.float32]:
    lonAbs = encoded % 100000000                           # last 8 digits
    lonSign = (encoded / 100000000) % 10                   # 9th digit from right
    latAbs = (encoded / 1000000000) % 10000000             # next 7 digits
    latSign = encoded / 100000000000000000                 # first digit
    
    lat = np.float32(latAbs / 100000.0 if latSign == 0 else -(latAbs / 100000.0))
    lon = np.float32(lonAbs / 100000.0 if lonSign == 0 else -(lonAbs / 100000.0))
    
    return (lat, lon)
    
```

```python
# {'Temperatures': [1625, 1769, 2027, 2045, 2299, 2694, 2954, 2805, 2666, 2333, 2120, 1868], 'Precipitation': [1054, 154, 0, 128, 0, 0, 0, 0, 58, 218, 1839, 12]}

location_map  = {}
climate_data : list[np.int16] = []
cursor = 0
with open("climate_optimized.json") as f:
    for line in f:
        row = json.loads(line)
        lat, lon = decodeCoordinates(row["Coordinate"])
        location_map[(lat,lon)] = len(climate_data)
        climate_data.append(np.int16(len(row["ClimateData"])))
        for year in row["ClimateData"].keys():
            climate_data.append(np.int16(year))
            row["ClimateData"][year]
            climate_data.extend(np.int16(x) for x in row["ClimateData"][year]["Temperatures"])
            climate_data.extend(np.int16(x) for x in row["ClimateData"][year]["Precipitation"])
        

    
```

```python
np_climate_data = np.array(climate_data)
```

```python
import pickle
with open("climate_compact2.pkl", "wb") as f:
    pickle.dump((location_map,np_climate_data), f)
```
